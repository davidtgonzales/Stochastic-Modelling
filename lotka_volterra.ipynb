This is a simulation of the **Lotka-Volterra System** where the population of a predator is dependent on the population of the prey and *vice versa*.

$let\;\;y_1\;=\;prey\;population$

$\;\;\;\;\;\;y_2\;=\;predator\;population$

$y_1\rightarrow2y_1$

$y_1+y_2\rightarrow2y_2$

$y_2\rightarrow\emptyset$

$where\;k_1,\;k_2,\;and\;k_3\;are\;the\;rate\;constants.$

## Deterministic Model

Define the rate of change of the predator and prey above using the rate law, assuming elementary reactions.

$\frac{dy_1}{dt}=k_1y_1-k_2y_1y_2$


$\frac{dy_2}{dt}=k_2y_1y_2-k_3y_2$

Import the required packages.

```python
%matplotlib inline
import matplotlib.pyplot as plt
```

Set the rate-constant values, initial concentrations, and time span. Concentrations variables are set as lists and will be appended with their new values after each step.

```python
k=[1.0,0.1,0.1]
y1=[4.0]
y2=[10.0]
maxt=100
```

Using Euler's method to solve for the progress of the two populations along time. The time variable is also set as a list and will progress along each time step (dt).

```python
t=[0]  #initial time
dt=0.1 #time step
while t[len(t)-1]<maxt:
    dy1=k[0]*y1[len(y1)-1]-k[1]*y1[len(y1)-1]*y2[len(y2)-1]
    dy2=k[1]*y1[len(y1)-1]*y2[len(y2)-1]-k[2]*y2[len(y2)-1]
    y1.append(float(y1[len(y1)-1]+dt*dy1))
    y2.append(float(y2[len(y2)-1]+dt*dy2))
    t.append(float(t[len(t)-1]+dt))
```

The results are plotted below for the time series and state-space.

```python
prey=plt.plot(t,y1,label='prey')
predator=plt.plot(t,y2,label='predator')
plt.xlabel('time',size=15)
plt.ylabel('population',size=15)
plt.legend()
plt.xlim(0,maxt)
plt.ylim(0,20)
plt.grid(b=True,which='major',color='grey',linestyle='--')
plt.show()
```

```python
statespace=plt.plot(y2,y1)
plt.xlabel('predator',size=15)
plt.ylabel('prey',size=15)
plt.grid(b=True, which='major', color='grey', linestyle='--')
plt.show()
```

## Stochastic Model

Import the required packages.

```python
import numpy as np
import random
```

Setup the Petri net and the maximum time of the simulation.

```python
pre=np.array([[1,0],[1,1],[0,1]])
post=np.array([[2,0],[0,2],[0,0]])
m=np.array([[50],[100]])
h=np.array([[],[],[]])
k=np.array([[1],[0.005],[0.6]])
maxt=25
```

The net effect matrix is calculated by $Post-Pre$.

```python
net=post-pre
```

Define the trajectory function. Include a way for the algorithm to handle race conditions and when the system becomes extinct.

```python
def Trajectory(maxt,net,m,h,k):
    t=np.array([0])
    while t[-1]<=maxt:
        h=np.append(h,[[k[:,-1][0]*m[:,-1][0]],
                       [k[:,-1][1]*m[:,-1][0]*m[:,-1][1]],
                       [k[:,-1][2]*m[:,-1][1]]],axis=1)
        h_total=sum(h[:,-1])
        if h_total==0:
            m=np.append(m,m[:,-1][:,None],axis=1)
            stept=1
            t=np.append(t,t[-1]+stept)
        else:
            prob=h[:,-1]/h_total
            interval=np.cumsum(prob)
            rxn=np.array([])
            u=random.uniform(0,1)
            prior=0
            for item in interval:
                if item>=u and prior==0:
                    rxn=np.append(rxn,1)
                elif item>=u and u>prior and prior!=0:
                    rxn=np.append(rxn,1)
                else:
                    rxn=np.append(rxn,0)
                prior=prior+item
            rxn=rxn.reshape(rxn.shape+(1,))
            m=np.append(m,m[:,-1][:,None]+np.dot(net.transpose(),rxn),axis=1)
            stept=random.expovariate(h_total)
            t=np.append(t,t[-1]+stept)
    return m,h,t
```

Run the trajectory function several times.

```python
m_matrix=[]
h_matrix=[]
t_matrix=[]
trials=1
for i in range(0,trials):
    m_result,h_result,t_result=Trajectory(maxt,net,m,h,k)
    m_matrix.append(m_result)
    h_matrix.append(h_result)
    t_matrix.append(t_result)
```

Plot the results.

```python
for i in range(0,trials):
    plt.plot(t_matrix[i],m_matrix[i][0,:],'blue',label='prey')
    plt.plot(t_matrix[i],m_matrix[i][1,:],'green',label='predator')
    if i==0:
        plt.legend(loc=2)
    
plt.xlabel('time (s)',size=15)
plt.ylabel('Concentration (M)',size=15)
plt.xlim(0,maxt)
plt.ylim(0,400)
plt.grid(b=True, which='major', color='grey', linestyle='--')
plt.show()
```

```python
for i in range(0,trials):
    plt.plot(m_matrix[i][1,:],m_matrix[i][0,:])

plt.xlabel('predator',size=15)
plt.ylabel('prey',size=15)
plt.grid(b=True, which='major', color='grey', linestyle='--')
plt.show()
```
